# Jupyter Notebook for responsible for create maize circRNA k-mer datasets

Raw data downloaded from: http://deepbiology.cn/crop/index.php/Home/Html/Download

```python
# Install kcounter lib 
#!pip install kcounter
```

```python
# Import libs
import pandas as pd
import numpy as np
import kcounter
pd.set_option('display.max_columns', None)
```

```python
# Create pandas dataframe based on raw data 
circ_df = pd.read_excel('maize_db.xlsx')

circ_df
```

## Create dataframe 1 mer

```python
# Create df - 1-mer
list = circ_df['seq']

df_kmer = pd.DataFrame()

for record in list:
  kmer_data = kcounter.count_kmers(record, 1, relative_frequencies=True)
  df_kmer = df_kmer.append(kmer_data, ignore_index=True)

df_kmer.fillna(0, inplace=True)
```

```python
# Concat raw data with k_mer data
complete_df = pd.concat([circ_df, df_kmer], axis=1)

complete_df
```

```python
# Export processed data
complete_df.to_excel('maize_1mer_dataset.xlsx', index=False)
```
